* Observaci√≥n: üòº

* Para el input de entrada de mols2grid se puede emplear un archivo SDF, UN DataFrame, o una lista de mol√©culas de RDKit
* Para mostrar la interface se puede insertar un pandas DataFrame como un input que contenga una columna de mol√©culas RDKit (especificando 'mol_col=...'), o SMILES string (especificando 'smiles_col=..')


We can also use a pandas DataFrame as input, containing a column of RDKit molecules (specified using `mol_col=...`) or SMILES strings (specified using `smiles_col=...`):

```python
#Emplear√© una columna de SMILES
mols2grid.display(df, smiles_col="smiles")
```

```python
#Ahora emplear√© SMILES
mols2grid.display(intento_final, smiles_col="smiles")
```
